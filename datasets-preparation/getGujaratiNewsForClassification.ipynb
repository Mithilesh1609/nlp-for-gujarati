```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import re
import pickle
import pandas as pd
```

```python
all_tech_headlines = []
```

```python
for i in range(1,39):
    url = 'https://gujarati.news18.com/tech/page-' + str(i) + '/'
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    html_doc = ''
    with urlopen(req) as response:
        for line in response:
            line = line.decode('utf-8')
            html_doc = html_doc + line.replace('\n','')
    soup = BeautifulSoup(html_doc, 'html.parser')
    blogs = soup.find_all('div',{'class':'blog-list-blog'})
    print(len(blogs))
    for blog in blogs:
        all_tech_headlines.append(blog.text)
```

```python
all_entertainment_headlines = []
```

```python
for i in range(1,82):
    url = 'https://gujarati.news18.com/entertainment/page-' + str(i) + '/'
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    html_doc = ''
    with urlopen(req) as response:
        for line in response:
            line = line.decode('utf-8')
            html_doc = html_doc + line.replace('\n','')
    soup = BeautifulSoup(html_doc, 'html.parser')
    blogs = soup.find_all('div',{'class':'blog-list-blog'})
    for blog in blogs:
        all_entertainment_headlines.append(blog.text)
    print(len(all_entertainment_headlines))
```

```python
entertainment_headlines = list(filter(lambda x: x != s, all_entertainment_headlines))
```

```python
entertainment_headlines = list(set(entertainment_headlines))
```

```python
len(entertainment_headlines)
```

```python
entertainment_headlines = list(set(entertainment_headlines)) 
business_headlines = list(set(business_headlines))
tech_headlines = list(set(tech_headlines))
```

```python
all_headlines = entertainment_headlines + business_headlines + tech_headlines
```

```python
len(all_headlines)
```

```python
all_labels = ['entertainment']*len(entertainment_headlines) + ['business']*len(business_headlines) + ['tech']*len(tech_headlines)
```
